```python
output_df = pd.DataFrame(columns = ['Chip','File1','File2','Intra-Hamming Distance','Percentage(%)'])
output2_df = pd.DataFrame(columns = ['Data1','Data2','Inter-Hamming Distance','Percentage(%)'])
chip_num = [1,3,5,7]
def file_name(x,chip):

  file_name = 'CHIP_' + str(chip) +'_Bitstrings_TV_'+ str(x) +'.txt'
  return file_name

def compare_files(file1,file2):
  count = 0

  with open(file1,'r') as f1, open(file2,'r') as f2:
    text1 = f1.read()
    text2 = f2.read()

    if len(text1)!=len(text2):
      raise ValueError("File Length Not Matching")

    for pos, (char1,char2) in enumerate(zip(text1,text2)):
      if char1!=char2:
        count+=1
  percentage = (count/len(text1))*100
  return count,percentage
```

```python
n =0
for chip in chip_num:
  for m in range(1,13):

    distance,percentage = compare_files(file_name(n,chip),file_name(m,chip))
    output_df = output_df.append({'Chip':'Chip '+str(int(chip)),'File1':"TV_"+str(n),'File2':"TV_"+str(m),'Intra-Hamming Distance':distance,'Percentage(%)':percentage},ignore_index=True)
```

```python
output_df.to_csv("output1.csv",index=False)
files.download("output1.csv")
display(output_df)
```

```python
pd.options.display.float_format = '{:6f}'.format
sliced_output_df = output_df[(output_df['Chip']=='Chip 5')]
std_dev = sliced_output_df['Percentage(%)'].std()
mean = sliced_output_df['Percentage(%)'].mean()
display(sliced_output_df)
```

```python
def inter_compare(i,j,lines):
  count = 0
  if len(lines[i])!=len(lines[j]):
    raise ValueError(i,j,"Line Data Length Not Matching")

  for pos, (char1,char2) in enumerate(zip(lines[i],lines[j])):
    if char1!=char2:
      count+=1
  percentage = (count/len(lines[i]))*100
  return count,percentage
```

```python
with open("SHDBitstrings.txt",'r') as f:
    lines = f.read().split("\n")

print(len(lines))
```

```python
for i in range(len(lines)-1):
  for j in range(i+1,len(lines)-1):
    distance,percentage = inter_compare(i,j,lines)
    output2_df = output2_df.append({'Data1':str(int(i)),'Data2':str(int(j)),'Inter-Hamming Distance':distance,'Percentage(%)':percentage},ignore_index=True)
```

```python
output2_df.to_csv("output2.csv",index=False)
files.download("output2.csv")
```
